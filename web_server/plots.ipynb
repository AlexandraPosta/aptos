```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../assets/Orientation.csv")
```

```python
df
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
plt.rcParams["figure.figsize"] = [12, 6]

df1 = pd.DataFrame(df, columns=['Aptos qw', 'Ref qw'])
df2 = pd.DataFrame(df, columns=['Aptos qx', 'Ref qx'])
df3 = pd.DataFrame(df, columns=['Aptos qy', 'Ref qy'])
df4 = pd.DataFrame(df, columns=['Aptos qz', 'Ref qz'])

df1.index = df['Time']
df2.index = df['Time']
df3.index = df['Time']
df4.index = df['Time']

df1.plot(fontsize=10, style={'Ref qw':'--', 'Aptos qw':'-'}, ax=axes[0,0], grid=True, xlabel='Time (s)', ylabel='Orientation w (rad/s)')
df2.plot(fontsize=10, style={'Ref qx':'--', 'Aptos qx':'-'}, ax=axes[0,1], grid=True, xlabel='Time (s)', ylabel='Orientation x (rad/s)')
df3.plot(fontsize=10, style={'Ref qy':'--', 'Aptos qy':'-'}, ax=axes[1,0], grid=True, xlabel='Time (s)', ylabel='Orientation y (rad/s)')
df4.plot(fontsize=10, style={'Ref qz':'--', 'Aptos qz':'-'}, ax=axes[1,1], grid=True, xlabel='Time (s)', ylabel='Orientation z (rad/s)')
```
